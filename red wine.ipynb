```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample DataFrame for demonstration
data = {
    'feature1': [1, 2, 3, 4, 5],
    'feature2': [10, 20, 30, 40, 50],
    'quality': [0, 1, 2, 0, 1]  # Example quality values
}
df = pd.DataFrame(data)

# Correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")

# Show the plot
plt.title('Correlation Matrix')
plt.show()
```

```python
# Set visual style
sns.set(style="whitegrid")
fig, axes = plt.subplots(4, 3, figsize=(15, 12))
fig.suptitle('Distribution of Features', fontsize=16)

# Plot histograms for each column
for i, column in enumerate(data.columns):
    sns.histplot(data[column], ax=axes[i//3, i%3], kde=True)
    axes[i//3, i%3].set_title(column)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample DataFrame
data = {
    'alcohol': [10.5, 11.0, 10.7, 12.5, 11.2],
    'quality': [0, 1, 2, 0, 1]
}
df = pd.DataFrame(data)

# Check DataFrame content
print(df)
print(df.columns)

# Create a scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='alcohol', y='quality', data=df)
plt.title('Alcohol Content vs Wine Quality')
plt.show()
```

```python
plt.figure(figsize=(5, 6))
sns.boxplot(x='quality', y='alcohol', data=df)
plt.title('Alcohol Content vs Quality')
plt.show()
```

```python
Analyze Relationships Between Key Features and Quality
```

```python
# Setting up plots for relationships between key features and quality
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Relationships between Key Features and Quality', fontsize=16)

# Alcohol vs Quality
sns.scatterplot(x="alcohol", y="quality", data=data, ax=axes[0, 0])
axes[0, 0].set_title("Alcohol vs Quality")

# Volatile Acidity vs Quality
sns.scatterplot(x="volatile acidity", y="quality", data=data, ax=axes[0, 1])
axes[0, 1].set_title("Volatile Acidity vs Quality")

# Citric Acid vs Quality
sns.scatterplot(x="citric acid", y="quality", data=data, ax=axes[1, 0])
axes[1, 0].set_title("Citric Acid vs Quality")

# Sulphates vs Quality
sns.scatterplot(x="sulphates", y="quality", data=data, ax=axes[1, 1])
axes[1, 1].set_title("Sulphates vs Quality")

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
